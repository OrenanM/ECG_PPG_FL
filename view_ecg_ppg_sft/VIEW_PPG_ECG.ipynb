```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import stft
import torchvision
```

```python
path = '../dataset/Generate_ECG_PPG/bidmc-ppg-and-respiration-dataset-1.0.0/bidmc_csv/bidmc_02_Signals.csv'
df = pd.read_csv(path)
```

```python
size_sample = 625
ecg = df[' II'].values[:size_sample]
t = df['Time [s]'].values[:size_sample]

T = 0.008
fs = 1/T

frequencies, times, Zxx = stft(ecg*1000, fs, nperseg=10)

plt.figure(figsize=(8, 4))
plt.plot(t, ecg*1000, color='blue')

plt.xlabel("Time (s)", fontsize=15)
plt.ylabel("Amplitude (mV)", fontsize=15)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.xlim(0, 5)
#plt.legend()
plt.grid(linestyle='--')
plt.savefig('ECG_time.pdf', bbox_inches='tight')
plt.show()

# Plotar o espectrograma da STFT
plt.figure(figsize=(8.4, 4))
plt.pcolormesh(times, frequencies, np.abs(Zxx), shading='gouraud')

cbar = plt.colorbar(label="Magnitude")
cbar.ax.yaxis.label.set_size(15)  # Aumentar o tamanho da fonte do rótulo da barra de cores
cbar.ax.tick_params(labelsize=13.5)  # Aumentar o tamanho da fonte dos números da barra de cores

plt.xlabel("Time (s)", fontsize=15)
plt.ylabel("Frequency (Hz)", fontsize=15)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.savefig('ECG_freq.pdf', bbox_inches='tight')
plt.show()
```
