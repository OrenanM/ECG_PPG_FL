```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import stft
import torchvision
```

```python
path = 'dataset/Generate_ECG_PPG/bidmc-ppg-and-respiration-dataset-1.0.0/bidmc_csv/bidmc_02_Signals.csv'
df = pd.read_csv(path)
```

```python
size_sample = 625
ecg = df[' II'].values[:size_sample]
t = df['Time [s]'].values[:size_sample]

T = 0.008
fs = 1/T

frequencies, times, Zxx = stft(ecg, fs, nperseg=10)


plt.figure(figsize=(10, 4))
plt.plot(t, ecg, label="Srie Temporal")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.title("Srie Temporal Aleatria")
plt.legend()
plt.grid()
plt.show()

# Plotar o espectrograma da STFT
plt.figure(figsize=(10, 5))
plt.pcolormesh(times, frequencies, np.abs(Zxx), shading='gouraud')
plt.colorbar(label="Magnitude")
plt.xlabel("Tempo (s)")
plt.ylabel("Frequncia (Hz)")
plt.title("Espectrograma da STFT")
plt.show()
```
